```python
import os, sys, datetime
import numpy as np
import re
```

```python
os.chdir(os.path.abspath(""))
os.getcwd()
```

```python
f = open("../result.txt", 'r', encoding="utf-8")
text = f.read()
```

```python
print(text[:1000])
```

```python
sample = text
sample = re.sub(r'\([^)]*\)', '', sample)
sample = re.sub(r'\[[^)]*\]', '', sample)
sample = re.sub("\s+", " ", sample)
sample = re.sub(r"[^가-힣0-9\.!?,a-zA-Z\s]+","", sample)
sample = re.sub(r'\.', '.\n', sample)
sample = re.sub(r'\!', '!\n', sample)
sample = re.sub(r'\?', '?\n', sample)
sample = re.sub(r'\n +', '\n', sample)

print(sample[:2000])
```

```python
result = open("../preprocessed.txt", "w+")
result.write(sample)
result.close()
```
